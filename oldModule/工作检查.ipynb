```python
import pandas as pd
file=r'D:\Projects\国土巡查\日期\202307\20230706\附件1-2022年上半年、下半年及2023年1-4月土地卫片执法实施“增违挂钩”冻结（挂账处理）补充耕地指标情况表.xlsx'
df=pd.read_excel(file,sheet_name='Sheet2',skiprows=3)
df.info()
```

```python
for name,df1 in df.groupby(['地级市']):
    print(f'---------------以下是{name}------------')
    print('a',f"{df1[df1['县']!='全市'].shape[0]}")
    print('b',f"{round(df1[df1['县']=='全市']['计划冻结指标数量22上'].sum(),2)}")
    print('c',f"{round(sum(df1[df1['县']=='全市']['计划冻结指标数量22下'].values+df1[df1['县']=='全市']['计划冻结指标数量23'].values),2)}")
    df_a=df1[(df1['县']!='全市') & (df1['计划冻结指标数量总计']>0)]
    df_b=df1[(df1['县']!='全市') & (df1['实际冻结指标数量']>0)]
    df_c=df1[(df1['县']!='全市') & (df1['实际冻结指标数量']==0)]
    df_f=df1[(df1['县']!='全市') & (df1['实际冻结指标数量']<df1['计划冻结指标数量总计']) & (df1['实际冻结指标数量']>0)]
    print('d',df_a.shape[0],f"{round(df_a['计划冻结指标数量总计'].sum(),2)}亩",f"无需冻结县区：{','.join(df_c['县'].values.tolist())}",f"实际比计划小：{','.join(df_f['县'].values.tolist())}",f"实际冻结：{df_b.shape[0]}个县区{round(df_b['实际冻结指标数量'].sum(),2)}亩")
    df_d=df1[(df1['县']!='全市') & (df1['挂账处理补充耕地指标总计']>0)]
    print('e',f"挂账冻结：{df_d.shape[0]}个县区{round(df_d['挂账处理补充耕地指标总计'].sum(),2)}亩")
```
